```python
import numpy as np
import cv2
from aux_functions import *

car_path = "training/vehicles/**/*.png"
non_car_path = "training/non-vehicles/**/*.png"
```

```python
cars = glob.glob(car_path, recursive=True)
non_cars = glob.glob(non_car_path, recursive=True)
```

```python
car_list = []
non_car_list = []

for i in range(5):
    i1 = random.randint(1, len(cars))
    i2  = random.randint(1, len(non_cars))
    car_list.append(cv2.imread(cars[i1]))
    non_car_list.append(cv2.imread(non_cars[i2]))
```

```python
fig = plt.figure(figsize=(20,10))

for ind in range(len(car_list)):
    plt.subplot(2,5,ind+1)
    plt.imshow(car_list[ind])
    plt.title('Random Car ' + str(ind+1))
    plt.subplot(2,5,ind+6)
    plt.imshow(non_car_list[ind])
    plt.title('Random Non-car ' + str(ind+1))
    
plt.show()
```
