```python
l = [1,2,3,4]

for num in l:
    print(num)
    loadpath = 'default/gg/out' + str(num) + '.p'
    outpath = 'video_out/gg/red' + str(num) + '.mp4'
    
    out1 = pickle.load(open(loadpath, 'rb'))
    
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(outpath, fourcc, 25.0, (1280, 720))

    for i in range(len(out1)):
        im = mpimg.imread('proframe_imgs/' + str(out1[i][0]) + '.jpg').astype(np.uint8)
        #im = np.stack((im[:,:,2], im[:,:,1], im[:,:,0]), axis=2)
        pixels = np.nonzero(out1[i][1])
        im[pixels] = np.array([0, 0, 255]).astype(np.uint8)

        out.write(im)

    out.release()
```
